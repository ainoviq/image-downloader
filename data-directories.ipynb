```python
import os
import pandas as pd
```

```python
dirs = []
def listdirs(rootdir):
    for it in os.scandir(rootdir):
        if it.is_dir():
            listdirs(it)
            dirs.append(it.path)
    return dirs
```

```python
directories = listdirs('./datasets')
directories
```

```python
dir_list = []
count_files =[]

for dir_ in directories:
    path, dirs, files = next(os.walk(dir_))
    if len(files) > 0:
        dir_list.append(dir_)
        count_files.append(len(files))
```

```python
dataframe = {'Directory': dir_list, 'Number of files': count_files}
df = pd.DataFrame(data=dataframe)
```

```python
df.to_csv('data-directories.csv', sep=',')
```
